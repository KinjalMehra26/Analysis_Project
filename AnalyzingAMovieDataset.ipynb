1. Data Cleaning and Preprocessing

```python
# Handle missing values in 'homepage' and 'tagline' by replacing them with 'Unknown'
df['homepage'].fillna('Unknown', inplace=True)
df['tagline'].fillna('Unknown', inplace=True)
```

```python
df
```

```python
# Fill missing values in 'runtime' with the median runtime
df['runtime'].fillna(df['runtime'].median(), inplace=True)
```

```python
# Remove duplicate entries
df.drop_duplicates(inplace=True)
```

```python
# Convert 'release_date' to datetime format
df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce')
```

```python
# Convert 'budget' and 'revenue' to numeric values
df['budget'] = pd.to_numeric(df['budget'], errors='coerce')
df['revenue'] = pd.to_numeric(df['revenue'], errors='coerce')
```

```python
df.info()
```

```python
# Extract 'release_year' from 'release_date'
df['release_year'] = df['release_date'].dt.year
```

2. Exploratory Data Analysis (EDA)

2.1 Analyze the distribution of movie genres

```python
# The 'genres' column contains genres as a JSON string. We'll need to extract the genres.
import ast
```

```python
# Function to extract genres from the JSON string
def get_genres(genre_str):
    genres = []
    try:
        genre_list = ast.literal_eval(genre_str)
        for genre in genre_list:
            genres.append(genre['name'])
    except:
        pass
    return genres
```

```python
# Apply the function to the 'genres' column
df['genres_list'] = df['genres'].apply(lambda x: get_genres(x))
```

```python
# Create a flat list of all genres
all_genres = [genre for sublist in df['genres_list'] for genre in sublist]
```

```python
# Count the frequency of each genre
from collections import Counter
genre_counts = Counter(all_genres)
```

```python
# Convert the genre counts to a DataFrame for visualization
genre_df = pd.DataFrame(genre_counts.items(), columns=['Genre', 'Count']).sort_values(by='Count', ascending=False)
```

```python
genre_df
```

```python
# Plot the genre distribution
plt.figure(figsize=(10, 6))
sns.barplot(x='Count', y='Genre', data=genre_df, palette='viridis')
plt.title('Distribution of Movie Genres')
plt.xlabel('Count')
plt.ylabel('Genre')
plt.show()
```
